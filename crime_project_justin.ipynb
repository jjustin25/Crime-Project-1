```python
import os
import pandas as pd
import glob
```

```python

indir="./cp16sb/"
outfile="../out/Concatenated.csv"

os.chdir(indir)
fileList=glob.glob("*.csv")
dfList=[]
for filename in fileList:
    print(filename)
    df=pd.read_csv(filename, header=None,encoding='cp1252')
    dfList.append(df)
concatDF=pd.concat(dfList, axis=0)
concatDF.to_csv(outfile, index=None)
```
